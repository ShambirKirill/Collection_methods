```python
# import of libraries
import requests
import json
from pprint import pprint
```

```python
url = 'https://api.nytimes.com/svc/search/v2/articlesearch.json'
category = 'Sports'
my_params = {'fq': 'news_desk:("' + category + '")',
             'api-key': ''}
```

```python
response = requests.get(url, params=my_params)
j_data = response.json()
```

```python
j_data
```

```python
for i in range(len(j_data.get('response').get('docs'))):
    print(j_data.get('response').get('docs')[i].get('headline').get('main'))
    print(j_data.get('response').get('docs')[i].get('web_url'))
```

```python
with open(f'apiNYT_categorynews.json', 'w') as f:
  json.dump(j_data, f, ensure_ascii=False)
```
